## Ridge Regression with Polynomial Features

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import Ridge
from warnings import filterwarnings
filterwarnings("ignore")
```

```python
def visualize_ridge_regression(x1, x2, alphas):
    plt.figure(figsize=(10, 6))
    plt.scatter(x1, x2, label='Datapoints', marker='*', color='purple')

    for alpha in alphas:
        model = Pipeline([
            ('poly_feats', PolynomialFeatures(degree=16)),
            ('ridge', Ridge(alpha=alpha))
        ])
        model.fit(x1, x2)
        
        x1_sorted = np.sort(x1, axis=0)
        preds = model.predict(x1_sorted)
        
        plt.plot(x1_sorted, preds, label=f'Alpha: {alpha}')

    plt.legend()
    plt.title('Ridge Regression with Polynomial Features')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.show()
```

```python
# Example usage
m = 100

x1 = 2 * np.random.rand(m, 1) - 2
x2 = 0.7 * x1 ** 2 - 2 * x1 + 3 + np.random.randn(m, 1)

alphas = [0,5, 200]
visualize_ridge_regression(x1, x2, alphas)
```

```python
from sklearn.linear_model import Lasso

def visualize_lasso_regression(x1, x2, alphas):
    plt.figure(figsize=(10, 6))
    plt.scatter(x1, x2, label='Datapoints', marker='*', color='purple')

    for alpha in alphas:
        model = Pipeline([
            ('poly_feats', PolynomialFeatures(degree=16)),
            ('lasso', Lasso(alpha=alpha))
        ])
        model.fit(x1, x2)
        
        x1_sorted = np.sort(x1, axis=0)
        preds = model.predict(x1_sorted)
        
        plt.plot(x1_sorted, preds, label=f'Alpha: {alpha}')

    plt.legend()
    plt.title('Lasso Regression with Polynomial Features')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.show()

# Example usage for Lasso
m = 100
x1 = 5 * np.random.rand(m, 1) - 2
x2 = 0.7 * x1 ** 2 - 2 * x1 + 3 + np.random.randn(m, 1)

lasso_alphas = [0.001, 0.01, 0.1]  # Specify Lasso alpha values
visualize_lasso_regression(x1, x2, lasso_alphas)
```

```python
from sklearn.linear_model import ElasticNet

def visualize_elastic_net_regression(x1, x2, alphas, l1_ratios):
    plt.figure(figsize=(7, 8))
    plt.scatter(x1, x2, label='Datapoints', marker='*', color='purple')

    for alpha in alphas:
        for l1_ratio in l1_ratios:
            model = Pipeline([
                ('poly_feats', PolynomialFeatures(degree=16)),
                ('elastic_net', ElasticNet(alpha=alpha, l1_ratio=l1_ratio))
            ])
            model.fit(x1, x2)

            x1_sorted = np.sort(x1, axis=0)
            preds = model.predict(x1_sorted)

            plt.plot(x1_sorted, preds, label=f'Alpha: {alpha}, L1 Ratio: {l1_ratio}')

    plt.legend(loc ='upper left',bbox_to_anchor=(1.05, 1))
    plt.title('Elastic Net Regression with Polynomial Features')
    plt.xlabel('X1')
    plt.ylabel('X2')
    plt.show()

# Example usage for Elastic Net
m = 100
x1 = 5 * np.random.rand(m, 1) - 2
x2 = 0.7 * x1 ** 2 - 2 * x1 + 3 + np.random.randn(m, 1)

elastic_net_alphas = [0.001, 0.01, 0.1]
elastic_net_l1_ratios = [0.1, 0.5, 0.9]
visualize_elastic_net_regression(x1, x2, elastic_net_alphas, elastic_net_l1_ratios)

```
